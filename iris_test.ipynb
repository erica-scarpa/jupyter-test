```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Display the first few rows of the dataset
iris.head()

# Calculate the mean of numerical columns
mean_values = iris.groupby("species").mean()
mean_values

# Plot the mean values
mean_values.plot(kind="bar", figsize=(10, 6), colormap="viridis")
plt.title("Mean Values by Species")
plt.xlabel("Species")
plt.ylabel("Mean Value")
plt.show()

from ipywidgets import interact

# Define an interactive function
@interact(feature=iris.columns[:-1])
def scatter_plot(feature):
    sns.scatterplot(data=iris, x=feature, y="sepal_length", hue="species")
    plt.title(f"Scatter Plot of {feature} vs Sepal Length")
    plt.show()
```

```python
from rdflib import Graph, Literal, URIRef

# Create a graph
g = Graph()

# Add triples to the graph
subject = URIRef("http://example.org/subject")
predicate = URIRef("http://example.org/predicate")
object = Literal("Hello, RDFLib!")
g.add((subject, predicate, object))

# Query the graph
query = """
    SELECT ?subject ?predicate ?object
    WHERE {
        ?subject ?predicate ?object
    }
"""
results = g.query(query)

# Print the results
for row in results:
    print(row)

# Serialize and print the graph in Turtle format
print(g.serialize(format="turtle").decode() if isinstance(g.serialize(format="turtle"), bytes) else g.serialize(format="turtle"))
```

```python
from rdflib import Graph, URIRef, Literal

# Create a graph
g = Graph()

# Define some URIs
book_uri = URIRef("http://example.org/book/1")
author_uri = URIRef("http://example.org/author/1")
genre_uri = URIRef("http://example.org/genre/fiction")

# Add triples to the graph
g.add((book_uri, URIRef("http://schema.org/name"), Literal("The Great Gatsby")))
g.add((book_uri, URIRef("http://schema.org/author"), author_uri))
g.add((author_uri, URIRef("http://schema.org/name"), Literal("F. Scott Fitzgerald")))
g.add((book_uri, URIRef("http://schema.org/genre"), genre_uri))
g.add((genre_uri, URIRef("http://schema.org/name"), Literal("Fiction")))

# Query the graph
query = """
    SELECT ?book ?author ?genre
    WHERE {
        ?book <http://schema.org/name> ?name .
        ?book <http://schema.org/author> ?author .
        ?book <http://schema.org/genre> ?genre .
    }
"""

results = g.query(query)

# Display the results
for row in results:
    print(row)
```
